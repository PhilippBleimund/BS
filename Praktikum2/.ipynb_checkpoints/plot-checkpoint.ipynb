```python
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
def load_data(input_path: str) -> pd.DataFrame:
    df = pd.read_csv(input_path, dtype=str)
    df['max_alloc']                      = pd.to_numeric(df['max_alloc'])
    df['first']             = pd.to_numeric(df['first'])
    df['next']              = pd.to_numeric(df['next'])
    df['best']            = pd.to_numeric(df['best'])
    df['worst']            = pd.to_numeric(df['worst'])
    return df
```

```python
def create_plot(df):
    fig, ax = plt.subplots()

    df_filtered = df[["first", "next", "best", "worst"]]
    
    for method in df_filtered.columns:
        ax.plot(df["max_alloc"], df_filtered[method], label=method)

    axins = zoomed_inset_axes(ax, 6, loc=1)

    ax.legend()
    plt.ylabel("num allocations")
    plt.xlabel("max_alloc")
    plt.title("vergleich allozierungsverfahren")
    ax.loglog()
    ax.grid(True)
    plt.show()
```

```python
def create_plot_new(df):
    fig, ax = plt.subplots()

    df_filtered = df[["first", "next", "best", "worst"]]

    # Hauptplot
    for method in df_filtered.columns:
        ax.plot(df["max_alloc"], df_filtered[method], label=method)

    ax.set_xscale("log")
    ax.set_yscale("log")

    ax.set_xlabel("max_alloc")
    ax.set_ylabel("num allocations")
    ax.set_title("Vergleich Allokierungsverfahren")
    ax.grid(True)
    ax.legend()

    # Zoom-Inset
    axins = zoomed_inset_axes(ax, zoom=6, loc="upper right")

    for method in df_filtered.columns:
        axins.plot(df["max_alloc"], df_filtered[method])

    # Bereich festlegen, der gezoomt werden soll
    axins.set_xlim(100, 1000)
    axins.set_ylim(1e3, 1e4)

    axins.set_xscale("log")
    axins.set_yscale("log")

    # Verbindung zwischen Hauptplot und Inset
    mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

    plt.show()
```

```python
INPUT_PATH = "~/BS/Praktikum2/data/benchmark_results20260623_095119.csv"

df = load_data(INPUT_PATH)
```

```python
create_plot(df)
```
